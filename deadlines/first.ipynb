```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
url = "https://raw.githubusercontent.com/s3drik/ai_uni/main/csv/automobile_data_nan.csv"

df = pd.read_csv(url)

print(df)
print(df.info)
```

```python
print("har bir qatordagi nan qiymatlar soni: ")
print(df.isnull().sum())
print("qaytarilgan qiymatlar soni")
print(df.duplicated().sum())
print("info:")
print(df.info())
```

```python
wh = df['wheel-base'].median()
le = df['length'].median()
et = df['engine-type'].mode()[0]

df.fillna({"wheel-base" : wh, "length" : le, "engine-type" : et}, inplace=True)
df.dropna(subset = ["horsepower", "average-mileage"], inplace=True)
#df["price"] = df["price"].fillna(method="ffill")
df["price"].ffill()
df.isnull().sum()
```

```python
cols = ["index", "wheel-base", "length", "horsepower", "average-mileage", "price"]
scaler = StandardScaler()
df_standard = df.copy()
df_standard[cols] = scaler.fit_transform(df[cols])
print(df_standard.head())
```
